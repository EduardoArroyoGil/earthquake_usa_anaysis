```python
#df_ec = wb.series.get("NY.GDP.MKTP.CD")
df_data_ec = wb.data.DataFrame('NY.GDP.MKTP.CD',wb.region.members('WLD'))

df_data_ec = df_data_ec.reset_index()
df_data_ec = pd.DataFrame(df_data_ec)
df_data_ec
```

```python
cols = ['YR' + str(i) for i in range(1960,2021)]
print(cols)
```

```python
df_melted = df_data_ec.melt(id_vars=['economy'],
                            value_vars=cols)
df_melted
```
